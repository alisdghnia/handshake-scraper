```python
# This is for each job card

def SearchJobCard(job_card):
    # job title
    position_type = job_card.find_element(By.XPATH, './/div[@class="style__job-type___oYHQq"]').text
    job_title = job_card.find_element(By.XPATH, './/div[@class="style__job-title___+ohfl"]').text
    company_name = job_card.find_element(By.XPATH, './/div[@class="style__job-detail___NfvjZ"]').text
    type_location = job_card.find_element(By.XPATH, './/div[@class="style__job-location___ZTuZq"]').text
    return job_title, position_type, company_name, type_location
```

```python
# list of job cards
job_cards = driver.find_elements(By.XPATH, '//div/a[@data-hook="jobs-card"]')
print(job_cards)


lst = []
# click to each of them
for job_card in job_cards:
    job_card.click()
    time.sleep(2) # wait for the job description to load

    # look into each job card
    # df = pd.DataFrame(SearchJobCard(job_card), columns=['job_title', 'position_type', 'company_name', 'type_location'])
    lst.append(SearchJobCard(job_card))
    lst
```

```python
df = pd.DataFrame(lst, columns=['job_title', 'position_type', 'company_name', 'type_location'])
```
