```python
lst = []
for index, row in df.iterrows():
    if row['Employment Type'] == 'Full-Time':
        if row['Payment Period'] == 'hour':
            stand_max_salary = row['Max salary'] * 40 * 52
            stand_min_salary = row['Min salary'] * 40 * 52
        elif row['Payment Period'] == 'month':
            stand_max_salary = row['Max salary'] * 12
            stand_min_salary = row['Min salary'] * 12
        else:
            stand_max_salary = row['Max salary']
            stand_min_salary = row['Min salary']
            
    elif row['Employment Type'] == 'Part-Time':
        if row['Payment Period'] == 'hour':
            stand_max_salary = row['Max salary'] * 20 * 52
            stand_min_salary = row['Min salary'] * 20 * 52
        elif row['Payment Period'] == 'month':
            stand_max_salary = row['Max salary'] * 12
            stand_min_salary = row['Min salary'] * 12
        else:
            stand_max_salary = row['Max salary']
            stand_min_salary = row['Min salary']
    else:
        stand_max_salary = row['Max salary']
        stand_min_salary = row['Min salary']

    lst.append([row['Job Title'], stand_min_salary, stand_max_salary])

df1 = pd.DataFrame(lst, columns=['Job Title', 'standardized_min_salary', 'standardized_max_salary'])
df1
```

```python
df_standardized = pd.merge(df, df1, left_index=True, right_index=True)
df_standardized
```
