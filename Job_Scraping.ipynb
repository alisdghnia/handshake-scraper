```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options


from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC


import time
from selenium.webdriver.common.keys import Keys

# data manipulate library
import pandas as pd

import csv

# address
from geopy.geocoders import Nominatim
```

```python
# set up header for job_info.csv file
header = ['job_type', 'job_title', 'company_name', 'location', 'application_deadline', 
          'posted_date', 'seasonal_role', 'estimated_salary', 'us-work-authorization',
          'number_of_employees', 'industry', 'headerquarter_location', 'post_url']
          

# write header to job_info.csv file
with open('job_info.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(header)

```

```python
chrome_options = Options()
chrome_options.add_experimental_option("debuggerAddress","localhost:8000")
driver = webdriver.Chrome(executable_path="chromedriver.exe", chrome_options=chrome_options)

# chrome.exe --remote-debugging-port=8000 --user-data-dir="D:\Chromedata"
```

```python
# This is for each job card

def SearchJobCard(job_card):
    try:
        # job type using WebDriverWait
        job_type = WebDriverWait(job_card, 10).until(EC.presence_of_element_located((By.XPATH, './/div[@class="style__job-type___oYHQq"]'))).text

        # job title using WebDriverWait
        job_title = WebDriverWait(job_card, 10).until(EC.presence_of_element_located((By.XPATH, './/div[@class="style__job-title___+ohfl"]'))).text

        # company name using WebDriverWait
        company_name = WebDriverWait(job_card, 10).until(EC.presence_of_element_located((By.XPATH, './/div[@class="style__job-detail___NfvjZ"]'))).text

        # location
        location = job_card.find_element(By.XPATH, './/div[@class="style__job-location___ZTuZq"]').text

        # application deadline
        try:
            application_deadline = driver.find_element(By.XPATH, '//div[@data-hook="application-deadline"]//div[2]').text
        except:
            application_deadline = None

        # posted date
        try:
            posted_date = driver.find_element(By.XPATH, '//div[@data-hook="posted-date"]//div[2]').text
        except:
            posted_date = None

        # seasonal role
        try:
            seasonal_role = driver.find_element(By.XPATH, '//div[@data-hook="seasonal-role"]//div[2]').text
        except:
            seasonal_role = None    

        # US work authorization
        try:
            us_work_authorization = driver.find_element(By.XPATH, '//div[@data-hook="us-work-authorization"]//div[2]').text
        except:
            us_work_authorization = None

        # estimated salary
        try:
            estimated_salary = driver.find_element(By.XPATH, '//div[@data-hook="estimated-pay"]//div[2]').text
        except:
            estimated_salary = None

        # number of employees
        try:
            number_of_employees = driver.find_element(By.XPATH, '//div[@class="style__content___sE2EK"]/parent::div//div[contains(text(),"Employees")]/following-sibling::div[1]').text
        except:
            number_of_employees = None

        # Industry
        try:
            industry = driver.find_element(By.XPATH, '//div[@class="style__content___sE2EK"]/parent::div//div[contains(text(),"Industry")]/following-sibling::div[1]').text
        except:
            industry = None

        # Headerquarter location
        try:
            headerquarter_location = driver.find_element(By.XPATH, '//div[@class="style__content___sE2EK"]/parent::div//div[contains(text(),"Headquarters")]/following-sibling::div[1]').text
        except:
            headerquarter_location = None

        # post URL
        try:
            post_url = driver.find_element(By.XPATH, '//h1[@class="style__job-title___P7PJV"]/a').get_attribute('href')
        except:
            post_url = None


        # print all the information on one line
        print(job_type, job_title, company_name, location, application_deadline, posted_date, seasonal_role, us_work_authorization, estimated_salary)
        
        # return a dictionary of all the information
        return {
            "job_type": job_type,
            "job_title": job_title,
            "company_name": company_name,
            "location": location,
            "application_deadline": application_deadline,
            "posted_date": posted_date,
            "seasonal_role": seasonal_role,
            "estimated_salary": estimated_salary,
            "us-work-authorization": us_work_authorization,
            "number_of_employees": number_of_employees,
            "industry": industry,
            "headerquarter_location": headerquarter_location,
            "post_url": post_url
        }

    except:
        print("Error")
        
        # return a dictionary of none
        return {
            "job_type": None,
            "job_title": None,
            "company_name": None,
            "location": None,
            "application_deadline": None,
            "posted_date": None,
            "seasonal_role": None,
            "estimated_salary": None,
            "us-work-authorization": None,
            "number_of_employees": None,
            "industry": None,
            "headerquarter_location": None,
            "post_url": None
        }

```

```python
def GoNextPage():
    # next page button
    next_page_button = driver.find_element(By.XPATH, '//button[@aria-label="next page"]')
    next_page_button.click()
    time.sleep(5)
```

```python
for i in range(1296):
    # list of job cards
    job_cards = driver.find_elements(By.XPATH, '//div/a[@data-hook="jobs-card"]')
    print(job_cards)

    # click to each of them
    for job_card in job_cards:
        job_card.click()
        
        # look into each job card and get the information
        job_info = SearchJobCard(job_card)
        
        # save to csv
        with open('job_info.csv', 'a', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(job_info.values())

    # click next page
    GoNextPage()
```
